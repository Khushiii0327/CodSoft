```python
# import libaries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read the data
movie = pd.read_csv("IMDb_Movies_India.csv", encoding="latin1")
movie
```

```python
movie.head()
```

```python
movie.info()
```

```python
movie.describe()
```

```python
# check missing value or NULL values    DATA CLEANING
movie.isnull().sum()
```

```python
movie.dropna(subset=["Rating"], inplace = True)
```

```python
movie.isnull().sum()
```

```python
movie.dropna(subset=['Actor 1','Actor 2','Actor 3','Director','Genre'],inplace=True)
```

```python
movie.isnull().sum()
```

```python
movie.head()
```

```python
# convert votes columns
movie['Votes'] = movie['Votes'].str.replace(',','').astype(int)
```

```python
# convert year columns
movie['Year'] = movie['Year'].str.strip('()').astype(int)
```

```python
# convert duration columns
movie['Duration'] = movie['Duration'].str.strip('min')
```

```python
movie['Duration'].fillna(movie['Duration'].median(),inplace=True)
```

```python
movie.isnull().sum()
```

```python
movie.head()
```

```python
NOW DATA IS CLEAN AND WE CAN DO THE ANALYSIS
```

```python
#TOP 10 MOVIES BASED ON RATINGS
top = movie.loc[movie['Rating'].sort_values(ascending=False)[:10].index]
top
```

```python
#TOP 10 MOVIES According to Year
sns.histplot(data=top, x="Year",hue="Rating", multiple="stack")
plt.title('Top 10 Movies', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Votes for top 10 movies per year
sns.barplot(x='Year',y='Votes',data=top)
plt.title("Votes By Year")
plt.show()
```

```python
#Genre distribution
genre = movie['Genre'].value_counts().reset_index()
genre.columns = ['Genre', 'Count']

# Select the top N genres (e.g., top 5)
top_genres = genre.head(5)
top_genres
```

```python
plt.figure(figsize=(4, 4))
plt.pie(top_genres['Count'], labels=top_genres['Genre'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Movie Genre Distribution', fontsize=16)
plt.axis('equal')
plt.show()
```

```python
# Group the data by director and calculate the average rating
director_avg_ratings = movie.groupby('Director')['Rating'].mean().reset_index()

director_avg_ratings = director_avg_ratings.sort_values(by='Rating', ascending=False)

top_director = director_avg_ratings.head(10)
top_director
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_director, x='Rating', y='Director', palette='viridis')

plt.title('Top Directors by Average Rating', fontsize=16)
plt.xlabel('Average Rating', fontsize=14)
plt.ylabel('Director', fontsize=14)

plt.show()
```

```python
#Relationship between the number of votes and movie ratings
plt.figure(figsize=(12, 6))
sns.lineplot(data=movie, x='Rating', y='Votes')

plt.title('Votes vs Rating', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Votes', fontsize=14)

plt.show()
```

```python
#Distribution of top actors by number of movie
actors = movie['Actor 1'].value_counts().reset_index()
actors.columns = ['Actor', 'MovieCount']

top_actors = actors.head(10)
top_actors
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_actors, x='MovieCount', y='Actor', orient='h')

# Set plot labels and title
plt.title('Top Actors by No of Movies', fontsize=16)
plt.xlabel('No of Movies', fontsize=14)
plt.ylabel('Actor', fontsize=14)

# Show the plot
plt.show()
```

```python
#Distribution of number of movie released every year
movie_count = movie['Year'].value_counts().reset_index()
movie_count.columns = ['Year', 'MovieCount']

movie_count = movie_count.sort_values(by='Year')
movie_count
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=movie_count, x='Year', y='MovieCount')

plt.title('Number of Movies Released Every Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Movies Released', fontsize=14)

plt.show()
```

```python
#Distribution of Movies with rating greater than 8 and votes greater than 10000
df= movie[(movie['Rating'] > 8) & (movie['Votes'] > 10000)]
df.head(10)
```

```python
plt.figure(figsize=(15, 6))
axl=sns.barplot(data=df,x='Name',y='Votes',hue='Rating',dodge=False,width=0.5,palette='muted')

axl.set_xticklabels(axl.get_xticklabels(), rotation=90, ha='right')
axl.legend(loc='upper right')
axl.set_xlabel('Movie Name')
axl.set_ylabel('Votes')
axl.set_title('Movies with rating greater than 8 and votes greater than 10000')

plt.show()
```

```python
#Distribution of Durations of movies by years
movie['Duration'] = movie['Duration'].astype(int)
movie['Year'] = movie['Year'].astype(int)
plt.figure(figsize=(12, 6))
sns.lineplot(data=movie,x='Year',y='Duration',errorbar=None)
plt.xlabel('Year')
plt.ylabel('Duration in minutes')
plt.title('Duration of movies by year')
plt.xticks(np.arange(1917,2023,5))
plt.show()
```

```python
#Distribution of Number of movies each genre
movie['Genre'] = movie['Genre'].str.split(',')

# Create a new DataFrame with one row for each genre
genre_df = movie.explode('Genre')
genre_df
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=genre_df, x='Genre', order=genre_df['Genre'].value_counts().index, palette='viridis')

plt.title('Number of Movies for Each Genre', fontsize=16)
plt.xlabel('Number of Movies', fontsize=14)
plt.ylabel('Genre', fontsize=14)
plt.xticks(rotation=90)

plt.show()
```

```python
#Distribution of Average rating of movies in each genre
average_rating = genre_df.groupby('Genre')['Rating'].mean().reset_index()
average_rating = average_rating.sort_values(by='Rating', ascending=False)
     

plt.figure(figsize=(15,5))
sns.barplot(data=average_rating,y="Rating",x='Genre',palette='coolwarm')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average rating of movies in each genre')
plt.xticks(rotation=90)

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

trans = movie.drop(['Name'],axis=1)
# Transform Director columns
trans['Director'] = label.fit_transform(movie['Director'])

# Transform Actors Columns
trans['Actor 1'] = label.fit_transform(movie['Actor 1'])
trans['Actor 2'] = label.fit_transform(movie['Actor 2'])
trans['Actor 3'] = label.fit_transform(movie['Actor 3'])

trans['Genre'] = label.fit_transform(movie['Genre'].apply(lambda x: ', '.join(x)))

trans.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()
sc = scale.fit_transform(trans)
sc = pd.DataFrame(sc,columns=trans.columns)

sc.head()
```

```python
# correlation
correl = trans.corr(numeric_only=True)
correl['Rating'].sort_values(ascending=False)
```

```python
sns.heatmap(correl,annot=False)
```

```python
#Model Building

# Import modeling Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
     

# put data except Rating data
X = trans.drop(['Rating'],axis=1)

# Put only Rating data
y = trans['Rating']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
X_test = np.array(X_test)
y_predict = model.predict(X_test)
y_predict
```

```python
print('R2 score: ',r2_score(y_test,y_predict))
print('Mean squared error: ',mean_squared_error(y_test,y_predict))
print('Mean absolute error: ',mean_absolute_error(y_test,y_predict))
```
